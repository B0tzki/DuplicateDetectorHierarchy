```python
from matplotlib import pyplot as plt
import numpy as np

def addlabels(x,y):
    sum_y = sum(y)
    for i in range(len(x)):
        # plt.text(x[i], y[i], y[i], ha = 'center')
        plt.text(x[i], y[i]//2, str(round(((y[i]/sum_y) * 100), 2)) + "%", ha = 'center', fontsize = 15)
    
def bar_charts(dictionary, filename, title, x_label, y_label):
  
  x = list(dictionary.keys())
  y = list(dictionary.values())

       
  # setting figure size by using figure() function 
  plt.figure(figsize = (10, 5))
    
  # making the bar chart on the data
  plt.bar(x, y)
    
  # calling the function to add value labels
  addlabels(x, y)
    
  # giving title to the plot
  plt.title(title, fontsize=20)
    
  # giving X and Y labels
  plt.xlabel(x_label, fontsize=17)
  plt.ylabel(y_label, fontsize=17)
  plt.xticks(size = 15)
  plt.yticks(size = 15)
    
  # visualizing the plot
 
  # plt.savefig(f'/drive/My Drive/rp_data/{filename}.pdf', bbox_inches='tight')
  plt.savefig(f'/drive/MyDrive/rp_data/plots_final/{filename}.pdf', bbox_inches='tight')
  # plt.savefig(f'{filename}.pdf', bbox_inches='tight')
  # plt.savefig(f'/drive/My Drive/rp_data/{filename}.pgf', bbox_inches='tight')
  plt.cla()
  plt.clf()
  plt.show()
  #This version also works if the file is downloaded - as a requirement
  # files.download(f'{filename}.pdf')

  # plt.plot()


def bar_charts_fraction(dictionary, filename, title, x_label, y_label):

  sorted_dictionary = sorted(dictionary.items())
  
  x = [str(round(a, 4)) for a, b in sorted_dictionary]
  # y = list(sorted_dictionary.values())
  y = [b for a, b in sorted_dictionary]

       
  # setting figure size by using figure() function 
  plt.figure(figsize = (15, 5))
    
  # making the bar chart on the data
  plt.bar(x, y)
    
  # calling the function to add value labels
  addlabels(x, y)
    
  # giving title to the plot
  plt.title(title, fontsize = 20)
    
  # giving X and Y labels
  plt.xlabel(x_label, fontsize = 17)
  plt.ylabel(y_label, fontsize = 17)
  plt.xticks(size = 15)
  plt.yticks(size = 15)
    
  # visualizing the plot
 
  # plt.savefig(f'/drive/My Drive/rp_data/{filename}.pdf', bbox_inches='tight')
  plt.savefig(f'/drive/MyDrive/rp_data/plots_final/{filename}.pdf', bbox_inches='tight')
  # plt.savefig(f'{filename}.pdf', bbox_inches='tight')
  # plt.savefig(f'/drive/My Drive/rp_data/{filename}.pgf', bbox_inches='tight')
  plt.cla()
  plt.clf()
  # plt.plot()
  plt.show()
  #This version also works if the file is downloaded - as a requirement
  # files.download(f'{filename}.pdf')
```

```python
def tags_hierarchy_count(df_questions, set_sedkgraph_tags):
    list_tags = df_questions['Tags']
    list_counts = []
    list_freq = []
    for list_post_tag in list_tags:
        count = 0
        for x in list_post_tag:
            if x in set_sedkgraph_tags:
                count = count + 1
        freq = count/len(list_post_tag)
        list_counts.append(count)
        list_freq.append(freq)
    df_questions['Count Tags Hier'] = list_counts
    df_questions['Freq Tags Hier'] = list_freq
    return df_questions


def so_tags_freq_histogram(df_questions):
    list_count_tags_hier = df_questions['Count Tags Hier']
    list_freq_tags_hier = df_questions['Freq Tags Hier']

    dict_count_tags_hier = dict()
    dict_freq_tags_hier = dict()
    for x in list_freq_tags_hier:
        if x in dict_freq_tags_hier:
            dict_freq_tags_hier[x] = dict_freq_tags_hier[x] + 1
        else: dict_freq_tags_hier[x] = 1

    for x in list_count_tags_hier:
        if x in dict_count_tags_hier:
            dict_count_tags_hier[x] = dict_count_tags_hier[x] + 1
        else:
            dict_count_tags_hier[x] = 1

    bar_charts_fraction(dict_freq_tags_hier, "posts_hier_freq_tags_per", "Hierarchy tags and their ratio distribution over SO posts", "The hierarchy tags ratio ", "No of SO posts")
    bar_charts(dict_count_tags_hier, "posts_hier_count_tags_per", "Hierarchy tags and their counting distribution over SO posts", "The hierarchy tags counting ", "No of SO posts")
    # bar_charts_fraction(dict_freq_tags_hier, "posts_hier_freq_tags_no", "Hierarchy tags and their ratio distribution over SO posts", "The hierarchy tags ratio ", "No of SO posts")
    # bar_charts(dict_count_tags_hier, "posts_hier_count_tags_no", "Hierarchy tags and their counting distribution over SO posts", "The hierarchy tags counting ", "No of SO posts")
```

```python
import statistics

def create_boxplot(data, filename, title):
  fig = plt.figure(figsize =(10, 7))  
  # Creating plot

  plt.boxplot(data)
  plt.title(title)
  # show plot
 
  plt.savefig(f'/drive/My Drive/rp_data/{filename}.pdf', bbox_inches='tight')
  # plt.savefig(f'{filename}.pdf', bbox_inches='tight')
  # plt.savefig(f'/drive/My Drive/rp_data/{filename}.pgf', bbox_inches='tight')
  plt.cla()
  plt.clf()
  plt.show()

def statistics_data(data):
  return statistics.mean(data), max(data), min(data), statistics.median(data)
```

```python
data_tags = df_test['Tags Count']
data_hier_tags = df_test['Count Tags Hier']
mean_tags, max_tags, min_tags, median_tags = statistics_data(data_tags)
mean_htags, max_htags, min_htags, median_htags = statistics_data(data_hier_tags)
create_boxplot(data_tags, "boxplot_tags_dist", "Boxplot for Tags Distribution")
create_boxplot(data_hier_tags, "boxplot_hier_tags_dist","Boxplot for Hierarchy Tags Distribution")

print('The statistics for Tags Count: \n Mean = {}, Max = {}, Min = {}, Median = {}'.format(mean_tags, max_tags, min_tags, median_tags))
print('The statistics for Hier Tags Count: \n Mean = {}, Max = {}, Min = {}, Median = {}'.format(mean_htags, max_htags, min_htags, median_htags))
```
